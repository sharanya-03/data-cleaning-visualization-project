## Sales Distribution Analysis

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=30)
plt.title('Sales Distribution')
plt.show()
```

The sales distribution is highly right-skewed. Most transactions involve lower sales amounts, while a small number of transactions contribute very high sales values.

## Sales by Category

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar')
plt.title('Total Sales by Category')
plt.ylabel('Sales')
plt.show()
```

The highest sales were generated by the Technology category, indicating strong customer demand in this segment.

## Profit by Region

```python
region_profit = df.groupby('Region')['Profit'].sum()

plt.figure(figsize=(8,5))
region_profit.plot(kind='bar')
plt.title('Profit by Region')
plt.ylabel('Profit')
plt.show()
```

The West region generated the highest profit, making it the best-performing region in terms of profitability.

## Sales vs Profit Relationship

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Sales', y='Profit')
plt.title('Sales vs Profit')
plt.show()
```

While higher sales often lead to higher profits, several high-sales transactions generated low or negative profit, suggesting the impact of discounts or operational costs.

## Correlation Analysis 

```python
plt.figure(figsize=(8,5))
sns.heatmap(
    df[['Sales','Quantity','Discount','Profit']].corr(),
    annot=True
)
plt.title('Correlation Heatmap')
plt.show()
```

Sales and Profit show a positive relationship, while Discount has a negative relationship with Profit.
